```python
#Importing the packages
import numpy as np
import matplotlib.pyplot as plt
```

```python
#We plot the Lagrangian polynomiala

x=np.linspace(-1,1,100)

#Linear elements
phi1=(x[-1]-x)/(x[-1]-x[0])
phi2=(x-x[0])/(x[-1]-x[0])

fig=plt.figure(figsize=(5,2))
ax=plt.subplot(111)
plt.plot(x,phi1,linewidth=1.5,color='red',label='$\phi_1$')
plt.plot(x,phi2,linewidth=1.5,color='blue',label='$\phi_2$')


# Graphical setup
plt.plot([x[-1], x[0]],[phi1[-1], phi1[0]],marker='o',linewidth=1.5,linestyle='none',markersize=10,color='red')
plt.plot([x[-1], x[0]],[phi2[-1], phi2[0]],marker='o',linewidth=1.5,linestyle='none',markersize=10,color='blue')

plt.xlabel('$x$', fontsize=12, labelpad=0.05)
plt.xlim([-1, 1])
plt.ylim([0, 1])
plt.yticks([0, 1])
plt.xticks([-1,1])
axes = plt.gca() #Getting the current axis
axes.spines[['top', 'right']].set_visible(False) #It works
plt.legend()
plt.show()
```

```python
#We plot the Lagrangian polynomials with order n
n=4
xj=np.linspace(-1,1,n)        #Collocation points
xx=np.linspace(-1,1,100)      #discretization of the element for the plot
#Recursive formulation

fig=plt.figure(figsize=(5,2))
legends=[['n=1'],['n=2'],['n=3'],['n=4'] ]

for jj in range(0,n) :
    phi=1
    for ii in range(0,n):
        if ii  != jj:
            phi= phi * (x-xj[ii])/(xj[jj]-xj[ii])

    
    ax=plt.subplot(111)
    plt.plot(x,phi,linewidth=1.5,label=legends[jj])
    

# Graphical setup
    
    
plt.xlabel('$x$', fontsize=12, labelpad=0.05)
plt.xlim([-1, 1])
#plt.ylim([0, 1])
#plt.yticks([-1, 1.5])
plt.xticks(xj)
axes = plt.gca() #Getting the current axis
axes.spines[['top', 'right']].set_visible(False) #It works
plt.legend()
plt.show()
```

```python

#We plot the Hermitia polynomials
import math

n=4
x=np.linspace(-1,1,100)      #discretization of the element for the plot
#Recursive formulation

fig=plt.figure(figsize=(5,2))

for N in range(0,n):
    H=np.zeros(100,)
    for m in range(math.floor(N/2)+1):
        H=H+math.factorial(N)*(((-1)**m)/(math.factorial(m)*math.factorial(N-2*m)))*(2*x)**(N-2*m)     
    ax=plt.subplot(111)
    plt.plot(x,H,linewidth=1.5,label=legends[N])
    
    
# Graphical setup
plt.xlabel('$x$', fontsize=12, labelpad=0.05)
plt.xlim([-1, 1])
#plt.ylim([0, 2])
axes = plt.gca() #Getting the current axis
axes.spines[['top', 'right']].set_visible(False) #It works
plt.legend(loc='lower right')
plt.show()


```
